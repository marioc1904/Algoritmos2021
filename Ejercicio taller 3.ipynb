```python
# Importación de paquetes especiales de Python

import numpy as np
import math
import matplotlib.pyplot as plt

#Definición de las masas molares y constantes

m_O2 = 5.31 * 10 ** -26       # Masa de una molecula de oxígeno, en kg
m_H2 = 3.35 * 10 ** -27       # Masa de una molecula de hidrógeno, en kg
m_N2 = 4.65 * 10 ** -26       # Masa de una molécula de nitrógeno, en kg
m_CO2 = 7.32 * 10 ** -26      # Masa de una molécula de dióxido de carbono, en kg

k_B = 1.38e-23                # Constante de Boltzamann en Joules/Kelvin

T = 300                       # Temperatura en Kelvin

#Definición del rango de velocidad y distribuciones

v = np.arange(10, 3000, 20)

n_O2 = 4 * math.pi * (m_O2 / (2 * math.pi * k_B * T)) ** (3 / 2) * (v**2) * np.exp(-m_O2 * v** 2 / (2 * k_B * T))

n_H2 = 4 * math.pi * (m_H2 / (2 * math.pi * k_B * T)) ** (3 / 2) * (v**2) * np.exp(-m_H2 * v**2 / (2 * k_B * T))

n_N2 = 4 * math.pi * (m_N2 / (2 * math.pi * k_B * T)) ** (3 / 2) * (v**2) * np.exp(-m_N2 * v**2 / (2 * k_B * T))

n_CO2 = 4 * math.pi * (m_CO2 / (2 * math.pi * k_B * T)) ** (3 / 2) * (v**2) * np.exp(-m_CO2 * v**2 / (2 * k_B * T))

#Distribuciones en una misma gráfica

plt.plot(v, n_O2, label = "Gas O2")

plt.plot(v, n_O2, label = "Gas O2")

plt.plot(v, n_O2, label = "Gas O2")




plt.xlabel('Velocidad (m/s)')                                   # Título en el eje horizontal
plt.ylabel('n(v)')                                              # Título en el eje vertical

plt.title('Distribución de Velocidad de Maxwell-Boltzmann')     # Título para la gráfica
plt.legend()                                                    # Leyenda
plt.show()                                                      # Mostrar gráfica
```

```python
# Importación de paquetes especiales de Python

import numpy as np
import math
import matplotlib.pyplot as plt

#Definición de las masas molares y constantes

m_O2 = 5.31 * 10 ** -26       # Masa de una molecula de oxígeno, en kg
m_H2 = 3.35 * 10 ** -27       # Masa de una molecula de hidrógeno, en kg
m_N2 = 4.65 * 10 ** -26       # Masa de una molécula de nitrógeno, en kg
m_CO2 = 7.32 * 10 ** -26      # Masa de una molécula de dióxido de carbono, en kg

k_B = 1.38e-23                # Constante de Boltzamann en Joules/Kelvin

T_baja = 80                   # Temperatura en Kelvin
T_media = 340
T_alta= 940
#Definición del rango de velocidad y distribuciones

v = np.arange(10, 3000, 20)

n_O2_Tbaja = 4 * math.pi * (m_O2 / (2 * math.pi * k_B * T_baja)) ** (3 / 2) * (v**2) * np.exp(-m_O2 * v** 2 / (2 * k_B * T_baja))

n_O2_Tmedia = 4 * math.pi * (m_O2 / (2 * math.pi * k_B * T_media)) ** (3 / 2) * (v**2) * np.exp(-m_O2 * v** 2 / (2 * k_B * T_media))

n_O2_Talta = 4 * math.pi * (m_O2 / (2 * math.pi * k_B * T_alta)) ** (3 / 2) * (v**2) * np.exp(-m_O2 * v** 2 / (2 * k_B * T_alta))



#Distribuciones en una misma gráfica

plt.plot(v, n_O2_Tbaja, label = "Gas O2 temperatura baja")

plt.plot(v, n_O2_Tmedia, label = "Gas O2 temperatura media")

plt.plot(v, n_O2_Talta, label = "Gas O2 temperatura alta")




plt.xlabel('Velocidad (m/s)')                                   # Título en el eje horizontal
plt.ylabel('n(v)')                                              # Título en el eje vertical

plt.title('Distribución de Velocidad de Maxwell-Boltzmann')     # Título para la gráfica
plt.legend()                                                    # Leyenda
plt.show()                                                      # Mostrar gráfica
```
